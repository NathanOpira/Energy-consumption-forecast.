```python
# Importing necessary libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Plotting style settings.
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)

# Loading the dataset.
# Parsing the 'date' column as datetime and store in a DataFrame.
df = pd.read_csv("../data/raw/energydata_complete.csv", parse_dates=["date"])

# Previewing the dataset.
print("Shape of dataset:", df.shape)
display(df.head())

# General info about the dataset.
print("\nDataset Info:")
df.info()

# Summary statistics.
print("\nDescriptive Statistics:")
display(df.describe())

# Checking for missing values.
print("\nMissing Values Per Column:")
print(df.isnull().sum())

# Setting 'date' as the datetime index and sort by time.
df.set_index("date", inplace=True)
df.sort_index(inplace=True)

# Visualizing the target variable (Appliances energy consumption over time).
plt.figure(figsize=(14, 6))
df["Appliances"].plot(title="Appliances Energy Consumption Over Time", color="blue")
plt.ylabel("Energy Consumption (Wh)")
plt.xlabel("Time")
plt.tight_layout()
plt.show()

# Correlation matrix (numerical features only)
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), cmap="coolwarm", center=0, annot=False, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()

# Distribution of the target variable
plt.figure(figsize=(10, 5))
sns.histplot(df["Appliances"], bins=50, kde=True, color='green')
plt.title("Distribution of Appliances Energy Consumption")
plt.xlabel("Appliances (Wh)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Time-of-day average energy usage
df['hour'] = df.index.hour
hourly_avg = df.groupby('hour')['Appliances'].mean()

plt.figure(figsize=(12, 5))
sns.lineplot(x=hourly_avg.index, y=hourly_avg.values)
plt.title("Average Energy Consumption by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Average Appliances Consumption (Wh)")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```
